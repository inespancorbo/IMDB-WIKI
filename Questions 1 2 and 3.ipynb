# 1. Find the The IMDB-WIKI dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
data["source"] = data["path"].apply(lambda x: x.split("/")[0])
data.head(10)
```

# 2. Plot the age distribution of this population, and determine the size of the bucket “15 to 25 years old”

```python
plt.figure(figsize = (10,5))
sns.distplot(data['age'])
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.distplot(data[data["source"]=="imdb_crop"]['age'], label="imdb_crop")
sns.distplot(data[data["source"]=="wiki_crop"]['age'], label="wiki_crop")
plt.title("Age Distribution")
plt.legend()
plt.show()
```

```python
# we can see some invalid or unlikely ages
print(data[data["age"] < 0].shape[0], data[data["age"] > 100].shape[0])
```

```python
data2 = data[data["age"].between(0, 100, inclusive = True)]
```

```python
plt.figure(figsize = (10,5))
sns.distplot(data2['age'])
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.distplot(data2[data2["source"]=="imdb_crop"]['age'], label="imdb_crop")
sns.distplot(data2[data2["source"]=="wiki_crop"]['age'], label="wiki_crop")
plt.title("Age Distribution")
plt.legend()
plt.show()
```

```python
bucket_num = data2[data2["age"].between(15, 25, inclusive = True)].shape[0]
bucket_num
```

```python
total_num = data2.shape[0]
total_num
```

```python
bucket_num/total_num
```
